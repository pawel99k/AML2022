```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
plt.rcParams['figure.figsize'] = (16,9)
plt.rcParams['font.size'] = 25
import matplotlib as mpl
```

```python
chi2=['$\chi^2$',0.880-0.01*(10/5-1)]
f_clasif=['Anova F-values',0.872-0.01*(15/5-1)]
mutual_info_clasif=['Mutual information',0.758-0.01*(20/5-1)]
AdaBoost=['AdaBoost',0.812]
LogisticRegression=['Logistic Regression',0.786]
RandomForest=['Random Forest', 0.858]
Boruta=['Boruta',0.882]
MCFS=['Monte Carlo',0.832]
FS_scores_artificial=pd.DataFrame([chi2,
                                   f_clasif,
                                   mutual_info_clasif,
                                   AdaBoost,
                                   LogisticRegression,
                                   RandomForest,
                                   Boruta,MCFS],
                                  columns=['Method','Score']).sort_values(by='Score',ascending=False)
```

```python
bars = sns.barplot(x='Method', y='Score', data=FS_scores_artificial)
mpl.rc('text', usetex = True)
for bar in bars.patches:
    bars.annotate(format(bar.get_height(), '.3f'),
                   (bar.get_x() + bar.get_width() / 2,
                    bar.get_height()-0.1), ha='center', va='center', xytext=(0, 8), rotation=90,
                   textcoords='offset points')
plt.ylim(0, 1)
plt.xlabel('Classifier')
plt.ylabel('Expected score')
plt.title('Artificial')
plt.xticks(rotation=45)
plt.show()
```
